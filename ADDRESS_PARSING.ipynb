```python
import pandas as pd #Import Libraries
import os
df = pd.read_excel (r"C:\Users\merit\Lavanya_Exercise1.xlsx")  #Read Excel
df
```

```python
df_addressparser=df.iloc[:,[0]]     #Extract Address Column
df_addressparser
```

```python
                                          #ADDRESS PARSER USING REGEX#
df_addressparser['Address']=df_addressparser['full_address'].str.extract('(.*?,[\s* | \d. |\d\s*}]*)')
df_addressparser['City']=df_addressparser['full_address'].str.extract('(\,+\s+[A-Za-z]+\s*[A-Za-z]{4}.)')
df_addressparser['State']=df_addressparser['full_address'].str.extract('(\,+\s+[A-Za-z]+[\, | \s]*[A-Za-z]{6})')
df_addressparser['Zip']=df_addressparser['full_address'].str.extract('(\s+\w*\s\d.?\d{0,4}\w{0,2})')
```

```python
df_addressparser #Display DataFrame
```

```python
df_addressparser.to_csv(os.getcwd()+'/ADDRESS_PARSER.csv',index=False) #Converting to CSV
```
